```python
from PIL import Image
import os
import tempfile

# Набір символів для ASCII-арту
ASCII_CHARS = ["@", "#", "S", "%", "?", "*", "+", ";", ":", ",", "."]

def resize_image(image, new_width=100):
    """Зміна розміру зображення з дотриманням співвідношення сторін."""
    width, height = image.size
    aspect_ratio = height / width / 1.65  # Корекція співвідношення
    new_height = int(new_width * aspect_ratio)
    resized_image = image.resize((new_width, new_height))
    return resized_image

def grayify(image):
    """Перетворення зображення у відтінки сірого."""
    return image.convert("L")

def pixels_to_ascii(image):
    """Конвертація пікселів у символи ASCII."""
    pixels = image.getdata()
    ascii_str = "".join([ASCII_CHARS[pixel // 25] for pixel in pixels])
    return ascii_str

def image_to_ascii(image_path, new_width=100):
    """Основна функція для перетворення зображення в ASCII-арт."""
    try:
        image = Image.open(image_path)
    except Exception as e:
        print(f"Не вдалося відкрити зображення: {e}")
        return

    # Перетворення зображення
    image = resize_image(image, new_width)
    gray_image = grayify(image)
    ascii_str = pixels_to_ascii(gray_image)

    # Форматування результату
    ascii_width = new_width
    ascii_art = "\n".join(
        [ascii_str[index : index + ascii_width] for index in range(0, len(ascii_str), ascii_width)]
    )
    return ascii_art

def save_ascii_art(ascii_art, filename="ascii_art.txt"):
    """Збереження ASCII-арту у файл."""
    try:
        with open(filename, "w") as f:
            f.write(ascii_art)
        print(f"ASCII-арт збережено у файл: {filename}")
    except OSError:
        print("Не вдалося записати файл у поточну директорію. Використовуємо тимчасовий файл.")
        with tempfile.NamedTemporaryFile(delete=False, suffix=".txt") as temp_file:
            temp_file.write(ascii_art.encode("utf-8"))
            print(f"ASCII-арт збережено тут: {temp_file.name}")

# Вказівка шляху до зображення
image_path = "/Users/mac/Documents/Univer/DF11.png"  # Замініть на шлях до вашого зображення
ascii_art = image_to_ascii(image_path)

# Виведення ASCII-арту
if ascii_art:
    print(ascii_art)
    save_ascii_art(ascii_art)


```
