```python
import matplotlib.pyplot as plt
```

```python
# Дані
роки = [1900, 1950, 1990, 2000, 2010, 2020, 2021, 2025]
дротові_телефони = [0.01, 0.05, 0.7, 1.0, 0.8, 0.8, 0.8, 0.8]  # у мільярдах
мобільні_телефони = [0, 0, 0.011, 0.74, 5.0, 0.1, 0.05, 0.04]  # у мільярдах (не смартфони)
смартфони = [0, 0, 0, 0, 0.3, 3.2, 3.5, 4.7]  # у мільярдах
```

```python
# Побудова графіка
plt.figure(figsize=(10, 6))
plt.plot(роки, дротові_телефони, marker='o', linestyle='-', color='blue', label='Дротові телефони')
plt.plot(роки, мобільні_телефони, marker='o', linestyle='-', color='green', label='Мобільні телефони (не смартфони)')
plt.plot(роки, смартфони, marker='o', linestyle='-', color='red', label='Смартфони')

# Додавання підписів
plt.title('Кількість телефонів у світі по роках', fontsize=16)
plt.xlabel('Рік', fontsize=10)
plt.ylabel('Кількість (млрд)', fontsize=10)
plt.xticks(роки)  # Відображення всіх років на осі X
plt.legend(fontsize=12)
plt.grid(True)

# Відображення графіка
plt.show()
```
